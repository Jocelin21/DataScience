```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import  SVC
import numpy as np
from sklearn.metrics import accuracy_score

df=pd.read_excel("Genshin Impact Revenue.xlsx")
df.dropna(inplace=True)
```

```python
df['Revenue Class']=np.where(df['Revenue']>df['Revenue'].mean(),'Good','Bad')
df
```

```python
df.info()
```

```python
X=df.iloc[:,[0,5,6,8]] #row,column
print(X)
y=df.loc[:,'Revenue Class']
print(y)
```

```python
svm = SVC(kernel='rbf', random_state=1, gamma=2.0, C=20.8)

#train the model
svm.fit(X, y)

#just use the same data to predict for now
y_pred=svm.predict(X)


print(accuracy_score(y, y_pred))
```
